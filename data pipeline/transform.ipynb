```python
import json
import pandas as pd
import os

# Load JSON bronze data
with open("btc_bronze_20260401_103401.json", "r") as f:
    raw = json.load(f)

df = pd.DataFrame(raw)

print(df.head())
print(f"Rows: {len(df)}")

# Convert types
df["date"] = pd.to_datetime(df["date"])
df["btc_close"] = pd.to_numeric(df["close"], errors="coerce")
df["btc_volume"] = pd.to_numeric(df["volume"], errors="coerce")

# Handle missing values
df = df.dropna(subset=["btc_close"])
df["btc_volume"] = df["btc_volume"].fillna(0)

# Sort BEFORE calculating difference
df = df.sort_values("date")

# Add gain/loss column
df["gain_loss"] = df["btc_close"].diff()

# fill first NaN
df["gain_loss"] = df["gain_loss"].fillna(0)

# Sort BEFORE calculating diff
df = df.sort_values("date")

# Gain/Loss
df["gain_loss"] = df["btc_close"].diff().fillna(0)

df["gain"] = df["gain_loss"] > 0

# Final table
silver = df[["date", "btc_close", "btc_volume", "gain_loss", "gain"]]


# Ensure folder exists
os.makedirs("data/silver", exist_ok=True)

# Save file
output_path = "data/silver/btc_daily_clean.csv"
silver.to_csv(output_path, index=False)

print(f"[SUCCESS] Saved to {output_path}")
```

```python
import json
import pandas as pd
import os

# ✅ INSERT YOUR BRONZE FILE HERE
input_file = "fng_bronze_20260401_104658.json"

# Load JSON
with open(input_file, "r") as f:
    raw = json.load(f)

# Create DataFrame
df = pd.DataFrame(raw)

# Debug
print(df.head())
print(f"Rows: {len(df)}")

# Correct columns
df["date"] = pd.to_datetime(df["date"])
df["fng_value"] = pd.to_numeric(df["value"], errors="coerce")
df["fng_classification"] = df["classification"]

# Handle missing values
df = df.dropna(subset=["fng_value"])

# Final silver table
silver = df[["date", "fng_value", "fng_classification"]].sort_values("date")

# Ensure folder exists
os.makedirs("data/silver", exist_ok=True)

# Save file
output_path = "data/silver/fng_daily_clean.csv"
silver.to_csv(output_path, index=False)

print(f"[SUCCESS] Saved to {output_path}")
```
