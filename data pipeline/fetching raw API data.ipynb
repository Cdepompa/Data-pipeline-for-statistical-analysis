```python
# raw_btc_fetch.py

# Import required libraries
import requests              
import json                  
from datetime import datetime  
import sys                   


# Function to fetch raw BTC data from Binance API
def fetch_raw_btc_data():
    url = "https://api.binance.com/api/v3/klines"  # Binance endpoint for candlestick (OHLCV) data
    
    # Parameters for the API request
    params = {
        "symbol": "BTCUSDT",   # Bitcoin price in USDT
        "interval": "1d",      # Daily data
        "limit": 365           # Last 365 days
    }

    try:
        # Send GET request to API with parameters and a timeout of 10 seconds
        response = requests.get(url, params=params, timeout=10)
        
        
        response.raise_for_status()
        
        
        return response.json()
# Handle any exceptions that occur during the request
    except requests.exceptions.RequestException as e:
        
        print(f"[ERROR] API request failed: {e}")
        sys.exit(1)  


# Function to save raw data into a JSON file
def save_raw_json(data):
   
    filename = f"btc_raw_{datetime.now().strftime('%Y%m%d_%H%M%S')}.json"

    try:
        # Open file in write mode
        with open(filename, "w") as file:
            # Dump raw data into file with indentation for readability
            json.dump(data, file, indent=2)

        print(f"[SUCCESS] Raw data saved to {filename}")

    except IOError as e:
        # Handle file writing errors (e.g. permission issues)
        print(f"[ERROR] Failed to write file: {e}")
        sys.exit(1)


# Main pipeline function (controls execution flow)
def run_pipeline():
    print("[INFO] Fetching RAW BTC data...")
    
    # Step 1: Fetch data from API
    data = fetch_raw_btc_data()

    print("[INFO] Saving RAW JSON...")
    
    # Step 2: Save raw data to file
    save_raw_json(data)


# Entry point of the script
# Ensures this runs only when script is executed directly (not imported)
if __name__ == "__main__":
    run_pipeline()
```

```python
# raw_fng_fetch.py

# Import required libraries
import requests              # Used to call the API
import json                  # Used to save data as JSON
from datetime import datetime  # Used to timestamp filenames
import sys                   # Used to exit program on failure


# Function to fetch raw Fear & Greed data from API
def fetch_raw_fng():
    # API endpoint
    url = "https://api.alternative.me/fng/"
    
    # Query parameters
    params = {
        "limit": 365,     # Get last 365 days of data
        "format": "json"  # Ensure response is JSON
    }

    try:
        # Send GET request to API
        response = requests.get(url, params=params, timeout=10)
        
        # Raise error if response status is not 200 (OK)
        response.raise_for_status()
        
        # Return raw JSON response (dictionary with "data" key)
        return response.json()

    except requests.exceptions.RequestException as e:
        # Catch any request-related issues (timeout, connection, bad response)
        print(f"[ERROR] API request failed: {e}")
        sys.exit(1)  # Stop execution if API fails


# Function to save raw JSON data to file
def save_raw_json(data):
    # Create unique filename using current date & time
    filename = f"fng_raw_{datetime.now().strftime('%Y%m%d_%H%M%S')}.json"

    try:
        # Open file in write mode
        with open(filename, "w") as f:
            # Save raw API response to file with indentation for readability
            json.dump(data, f, indent=2)

        print(f"[SUCCESS] Raw data saved to {filename}")

    except IOError as e:
        # Catch file writing issues (permissions, disk issues, etc.)
        print(f"[ERROR] Failed to write file: {e}")
        sys.exit(1)


# Main pipeline function (controls execution flow)
def run_pipeline():
    print("[INFO] Fetching RAW Fear & Greed data...")
    
    # Step 1: Fetch raw data from API
    data = fetch_raw_fng()

    print("[INFO] Saving RAW JSON...")
    
    # Step 2: Save raw data to file
    save_raw_json(data)


# Entry point of the script
# Ensures code runs only when script is executed directly
if __name__ == "__main__":
    run_pipeline()
```
